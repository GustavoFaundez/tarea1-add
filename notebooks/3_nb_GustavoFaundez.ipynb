```python
import calendar 
  
def findDay(param): 
    try:
        mes = 0
        if param['mes'] == 'ene':
            mes = 1
        elif param['mes'] == 'feb':
            mes = 2
        elif param['mes'] == 'mar':
            mes = 3
        elif param['mes'] == 'abr':
            mes = 4
        elif param['mes'] == 'may':
            mes = 5
        elif param['mes'] == 'jun':
            mes = 6
        elif param['mes'] == 'jul':
            mes = 7
        elif param['mes'] == 'ago':
            mes = 8
        elif param['mes'] == 'sep':
            mes = 9
        elif param['mes'] == 'oct':
            mes = 10
        elif param['mes'] == 'nov':
            mes = 11
        elif param['mes'] == 'dic':
            mes = 12

        total_fecha = str(int(param["dia"]))+' '+str(mes)+' '+str(param['year'])
        #print(total_fecha)

        day, month, year = (int(i) for i in total_fecha.split(' '))     
        dayNumber = calendar.weekday(year, month, day) 
        days =["Lunes", "Martes", "Miércoles", "Jueves", 
                             "Viernes", "Sábado", "Domingo"] 
        return (days[dayNumber]) 
    except ValueError:
        pass

#print(findDay(date)) 

df_sii["dia_de_la_semana"] = df_sii.apply(findDay, axis=1)
df_sii
```
